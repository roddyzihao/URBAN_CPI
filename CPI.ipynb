## URBAN Inflation Beijing vs. Shanxi by CPI

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_Beijing = pd.read_excel(r'/Users/zihao/Desktop/food/CPI/Beijing CPI.xlsx')
df_Shanxi = pd.read_excel(r'/Users/zihao/Desktop/food/CPI/ShanXi CPI.xlsx')
```

```python
df_Beijing.set_index('YEAR', inplace=True)
df_Beijing_percentage=((df_Beijing - df_Beijing.shift()) / df_Beijing.shift())
df_Shanxi.set_index('YEAR', inplace=True)
df_Shanxi_percentage=((df_Shanxi - df_Shanxi.shift()) / df_Shanxi.shift())
```

```python
df_Inflation = pd.concat([df_Beijing_percentage, df_Shanxi_percentage], axis=1)
df_Inflation.columns = ['Beijing', 'Shanxi']
```

```python
fig, ax = plt.subplots()

df_Inflation.plot(ax=ax, legend=True)

ax.set_title("Urban Weighted Average Inflation Rate")
ax.set_ylabel("Inflation")
```
